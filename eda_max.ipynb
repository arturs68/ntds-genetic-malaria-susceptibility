```python
import pandas as pd
import glob
```

```python
def load_df(path, encoding="utf-8"):
    loaded_pd = pd.read_csv(path, sep="\t", encoding=encoding)
    pd_name = path.split("/")[-1].split(".")[0]
    return pd_name, loaded_pd
```

```python
paths_expression = glob.glob("data/expression data/*.txt")
```

```python
expression_data = []
for path in paths_expression:
    expression_data.append(load_df(path))
```

```python
paths_other = glob.glob("data/*.txt")
aligner_path = "data/phenotypes_id_aligner.txt"
other_data = []
for path in paths_other:
    if path == aligner_path:
        other_data.append(load_df(path, encoding="latin"))
    else:
        other_data.append(load_df(path))
```

## We have loaded the dataframes to lists

Each list contains tuples (name, dataframe)
- expression_data: tables with gene expressions
- other_data: all the remaining tables

```python
Immune = other_data[3][1][other_data[3][1]['Category'] == 'Immune']
```

```python
sorted_phenotype = Immune.sort_values('Strains', ascending = False)
sorted_phenotype.head(50)
```

```python
sorted_phenotype.dtypes
```

```python
phenotypes_interest =  sorted_phenotype[(sorted_phenotype['Shown_pheno'].str.contains('influenza') |
                                         sorted_phenotype['Shown_pheno'].str.contains('Malaria')) &
                                       (sorted_phenotype['Strains'] >= 50)]
```

```python
phenotypes_interest.head(50)
```

```python
phenotypes_interest.shape
```

```python
phenotypes_dataframe = phenotypes_interest[['PhenoID', 'Shown_pheno']]
```

```python
phenotypes_dataframe.head()
```

```python
def to_disease(string):
    if 'Malaria' in string:
        return 'Malaria'
    else:
        return 'Influenza'
```

```python
phenotypes_dataframe['Disease'] = phenotypes_dataframe['Shown_pheno'].copy().apply(lambda string: to_disease(string))
```
